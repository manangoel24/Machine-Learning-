```python
import pandas as pd
from sklearn.linear_model import Ridge

df = pd.read_csv("pizza_data.csv")

X = df[[
    "Company",
    "Pizza Name",
    "Type",
    "Diameter"
]]

X = pd.get_dummies(X)

y = df["Price"]

model = Ridge()
model.fit(X, y)

print("Ridge Regression model trained successfully")
```

```python
print("R² Score:", model.score(X, y))
```
