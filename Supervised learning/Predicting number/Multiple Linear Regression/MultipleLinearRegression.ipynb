```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Read file
df = pd.read_csv("pizza_data.csv")

# Features and target
X = df[[
    "Company",
    "Pizza Name",
    "Type",
    "Diameter"
]]

# Convert categorical columns into numbers
X = pd.get_dummies(X)

y = df["Price"]

# Train model
model = LinearRegression()
model.fit(X, y)

# R² Score
r2 = model.score(X, y)

print("Multiple Linear Regression")
print("R² Score:", r2)
print("----------------------")
```
