```python
import pandas as pd
from sklearn.linear_model import LinearRegression

df = pd.read_csv("pizza_data.csv")

X = df[["Diameter"]].values
y = df["Price"]

model = LinearRegression()
model.fit(X, y)

def PizzaPrice(diameter):
    pred = model.predict([[diameter]])
    print("Predicted Price:",round(pred[0],2))
```

```python
PizzaPrice(12)
```

```python
print("R Score:" , model.score(X,y)) 
```

```python
import matplotlib.pyplot as plt

# Scatter plot of actual data
plt.scatter(X, y, color='blue', label='Actual Data')

# Regression line
plt.plot(X, model.predict(X), color='red', label='Regression Line')

# Labels and title
plt.xlabel("Diameter")
plt.ylabel("Price")
plt.title("Pizza Price vs Diameter")
plt.legend()

# Show graph
plt.show()
```
