```python
import pandas as pd
from sklearn.svm import SVC

df = pd.read_csv("pizza_data.csv")

X = df[["Diameter"]].values
y = df["Bought "]

model = SVC(probability=True)
model.fit(X, y)

def PizzaPrediction(diameter):
    pred = model.predict([[diameter]])

    if pred[0] == 1:
        print("Customer will buy the pizza")
    else:
        print("Customer will not buy the pizza")

def PizzaProbability(diameter):
    prob = model.predict_proba([[diameter]])
    print("Buying Probability:", prob[0][1])

PizzaPrediction(14)
PizzaProbability(14)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Scatter plot
plt.scatter(X, y, color="blue", label="Actual Data")

# Generate smooth x-values
x_values = np.linspace(X.min(), X.max(), 200).reshape(-1, 1)

# Predict probabilities
y_prob = model.predict_proba(x_values)[:, 1]

# Plot probability curve
plt.plot(x_values, y_prob, color="red", linewidth=2, label="Buying Probability")

plt.xlabel("Diameter")
plt.ylabel("Probability of Buying")
plt.title("SVM Probability Curve")
plt.legend()
plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.scatter(X, y, color="blue", label="Training Data")

x_values = np.linspace(X.min(), X.max(), 200).reshape(-1, 1)
y_pred = model.predict(x_values)

plt.plot(x_values, y_pred, color="red", linewidth=2, label="Decision Boundary")

plt.xlabel("Diameter")
plt.ylabel("Bought")
plt.title("SVM Classification")
plt.legend()
plt.grid(True)

plt.show()
```
