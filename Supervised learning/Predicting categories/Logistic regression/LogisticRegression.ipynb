```python
import pandas as pd
from sklearn.linear_model import LogisticRegression

# Read file
df = pd.read_csv("pizza_data.csv")

# Feature and target
X = df[["Diameter"]].values
y = df["Bought"]

# Train
model = LogisticRegression()
model.fit(X, y)

# Function
def PizzaPrediction(diameter):
    pred = model.predict([[diameter]])
    prob = model.predict_proba([[diameter]])
    print("Prediction:", pred[0])
    print("Probability of Buying:", round(prob[0][1] * 100, 2), "%")
    
```

```python
PizzaPrediction(5)
```

```python
print("Score:", round(model.score(X, y) * 100, 2), "%")
```
