```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB

# Read file
df = pd.read_csv("pizza_data.csv")

# Features and target
X = df[[
    "Company",
    "Pizza Name",
    "Type",
    "Diameter"
]]

# Convert categorical columns into numbers
X = pd.get_dummies(X)

y = df["Bought"]

# Train model
model = GaussianNB()
model.fit(X, y)

# Accuracy
accuracy = model.score(X, y)

print("Naive Bayes")
print("Accuracy:", accuracy)
print("----------------------")
```
