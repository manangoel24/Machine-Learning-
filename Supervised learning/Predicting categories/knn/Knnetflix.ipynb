```python
import pandas as pd
from sklearn.neighbors import NearestNeighbors

movies = pd.read_csv("Netflix_Dataset_Movie.csv")
ratings = pd.read_csv("Netflix_Dataset_Rating.csv")

df = pd.merge(ratings, movies, on="Movie_ID")

movie_table = df.pivot_table(
    index="Name",
    columns="User_ID",
    values="Rating"
)

movie_table = movie_table.fillna(0)

model = NearestNeighbors(n_neighbors=6, metric="cosine")

model.fit(movie_table)

def RecommendMovies(movie_name):

    distances, indices = model.kneighbors(
        [movie_table.loc[movie_name]]
    )

    print("\nMovies similar to:", movie_name)
    print("--------------------------------")

    for i in indices[0][1:]:
        print(movie_table.index[i])
```

```python
movie_names = pd.DataFrame(movie_table.index)

movie_names.sample(50)
```

```python
RecommendMovies("Vampires")
```
