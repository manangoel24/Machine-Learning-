```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

df = pd.read_csv("football_data.csv")

e = ["Weather", "Temperature", "Humidity", "Wind"]

mappings = {
    "Weather": {"Sunny": 0, "Cloudy": 1, "Rain": 2},
    "Temperature": {"Cool": 0, "Mild": 1, "Hot": 2},
    "Humidity": {"Normal": 0, "High": 1},
    "Wind": {"Weak": 0, "Strong": 1},
    "PlayFootball": {"No": 0, "Yes": 1}
}

for key, values in mappings.items():
    df[key] = df[key].map(values)

X = df[e].values
y = df["PlayFootball"]

model = DecisionTreeClassifier()
model.fit(X, y)

def PlayFootball(e):
    pred = model.predict([e])

    if pred[0] == 1:
        print("Prediction: Yes, you can play football")
    else:
        print("Prediction: No, you should not play football")

user_input = []

for key in e:
    value = input(f"Enter {key}: ")
    user_input.append(mappings[key][value])

PlayFootball(user_input)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))

plot_tree(
    model,
    feature_names=e,
    class_names=["No", "Yes"],
    filled=True,
    rounded=True,
    fontsize=10
)

plt.show()
```
