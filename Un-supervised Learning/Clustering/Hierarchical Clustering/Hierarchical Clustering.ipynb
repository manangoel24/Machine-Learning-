# Hierarchical Clustering

## Ward Linkage

```python
import pandas as pd
from sklearn.cluster import AgglomerativeClustering

# Read file
df = pd.read_csv("customer_dataset.csv")

# Features
e = [
    "Spending",
    "Visits",
    "AverageBill",
    "ItemsBought"
]

# Select features
X = df[e]

# Create model
model = AgglomerativeClustering(n_clusters=3)

# Train and predict
df["Cluster"] = model.fit_predict(X)

print(df) 
```

```python
print(df.groupby("Cluster")[e].mean()) #to know what each cluster represent
```

### Ward Linkage
Created the most balanced clusters: Low-value, Medium-value, and High-value customers.

Cluster averages are clearly separated (₹2.3k → ₹9.9k → ₹27k spending).

Ward tries to minimize variance within each cluster, so clusters are compact and well-formed.

## Single Linkage 

```python
import pandas as pd
from sklearn.cluster import AgglomerativeClustering

# Read file
df = pd.read_csv("customer_dataset.csv")

# Features
e = [
    "Spending",
    "Visits",
    "AverageBill",
    "ItemsBought"
]

# Select features
X = df[e]

# Create model
model = AgglomerativeClustering(
    n_clusters=3,
    linkage="single"
)
# Train and predict
df["Cluster"] = model.fit_predict(X)

print(df) 
```

```python
print(df.groupby("Cluster")[e].mean()) #to know what each cluster represent
```

### Single Linkage
Almost all customers were grouped into one large cluster (Cluster 0).
    
Only a few extreme customers formed Clusters 1 and 2.
    
This is called the chaining effect: clusters get connected through nearest neighbors, creating one long cluster.

## Complete Linkage 

```python
import pandas as pd
from sklearn.cluster import AgglomerativeClustering

# Read file
df = pd.read_csv("customer_dataset.csv")

# Features
e = [
    "Spending",
    "Visits",
    "AverageBill",
    "ItemsBought"
]

# Select features
X = df[e]

# Create model
model = AgglomerativeClustering(
    n_clusters=3,
    linkage="complete"
)
# Train and predict
df["Cluster"] = model.fit_predict(X)

print(df) 
```

```python
print(df.groupby("Cluster")[e].mean()) #to know what each cluster represent
```

## Complete Linkage
Produced one large low/medium-spending cluster and two smaller high-spending clusters.

High-value customers were split because complete linkage focuses on the farthest distance between points.

Creates tighter, more compact clusters than single linkage.

## Average Linkage 

```python
import pandas as pd
from sklearn.cluster import AgglomerativeClustering

# Read file
df = pd.read_csv("customer_dataset.csv")

# Features
e = [
    "Spending",
    "Visits",
    "AverageBill",
    "ItemsBought"
]

# Select features
X = df[e]

# Create model
model = AgglomerativeClustering(
    n_clusters=3,
    linkage="average"
)
# Train and predict
df["Cluster"] = model.fit_predict(X)

print(df) 
```

```python
print(df.groupby("Cluster")[e].mean()) #to know what each cluster represent
```

## Average Linkage
Produced a clear separation between low-value and high-value customers.
    
Similar to complete linkage, but less strict because it uses the average distance between clusters.
    
Resulting clusters are reasonably balanced and less affected by outliers
