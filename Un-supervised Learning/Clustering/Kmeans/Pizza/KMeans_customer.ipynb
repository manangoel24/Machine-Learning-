```python
import os
os.environ["OMP_NUM_THREADS"] = "1"

import pandas as pd
from sklearn.cluster import KMeans


df = pd.read_csv("customer_dataset.csv")

# Features
features = [
    "Spending",
    "Visits",
    "AverageBill",
    "ItemsBought"
]

# Select features
X = df[features]

# Create model
model = KMeans(n_clusters=4, random_state=42)

# Train model
model.fit(X)

# Add cluster result to dataset
df["Cluster"] = model.labels_

# Show result
print(df)
```

```python
print(df.groupby("Cluster")[features].mean()) #to know what each cluster represent 
```

```python
# Elbow Method

for k in range(1, 11):

    model = KMeans(n_clusters=k, random_state=42)

    model.fit(X)

    print("Clusters:", k, "Inertia:", model.inertia_) #to know what value of n_clusters should be 
```
