```python
import pandas as pd
from sklearn.decomposition import PCA

df = pd.read_csv("pizza_data.csv")

features = [
    "Company",
    "Pizza Name",
    "Type",
    "Diameter"
]

X = df[features]

X = pd.get_dummies(X)

model = PCA(n_components=2)

X_pca = model.fit_transform(X)

df["PC1"] = X_pca[:, 0]
df["PC2"] = X_pca[:, 1]

print(df)
```

```python
print("Explained Variance Ratio:", model.explained_variance_ratio_)
print("Total Information Kept:", model.explained_variance_ratio_.sum())
```
